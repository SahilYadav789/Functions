**Q6. Explain the concept of generators in Python and how they are defined?**

Generators in python are special type of iterable that produce values lazily, one at a time, rather than storing all values in memory simultaniously. This makes them highly memory efficient.

Generators are difined by def keyword just like other functions are defined but it contains yield keyword insted of return keyword.

```python
def even_numbers_generator(n):
    num=0
    while True:
      yield num
      num+=2
      if num>n:
        break

limit=int(input("Enter the limit: "))
for num in even_numbers_generator(limit):
    print(num)
```

**Q7. What are the advantages of using generators over regular functions?**

The advantages of using generators over regular functions are:

1. Memory efficiency: Generators yields values one at a time instead of generating and storing an entire sequence in memory. This is crucial when working on large datasets or infinte sequences.

2. Improved readability and maintainablity: Generators can encpsulate the iteration logic within a function using yield, making the code more concise and easier to understand.

3. State retention: Regular function executes and return a single value before exiting, generators retain their state between yeild calls allowing to resume from where they left off.

```python
def fibonacci_generator(n):
    a, b = 0, 1
    for i in range(10):
        yield a
        a, b = b, a + b

limit=100
for num in fibonacci_generator(limit):
    print(num)
```

**Q10. What is the diffenece between 'map( )','reduce( ), and 'filter( )' functions in Python?**

The map( ) function applies a given function to each item in an iterable and returns an iterator of the results. It transforms each element individually.

Example:- Adding each numbers in a list.

The filter( ) function constructs an iterator from the elements of an iterable for which a given function returns True. It ios used for selecting elements based on a condition.

Example:- Filtering out odd numbers from a list.

The reduce( ) function applies a rolling computation to sequential pairs of elements in an iterable, reducing reducing it to a single cumulative result. It is used for aggregating elements. It is found in functool library.

Example:- Doing arithmatical operations of all numbers in a list

```python
# Example map()

num1=[1,2,3,4,5]
print("OLD LIST-",num1)
square=map(lambda x: x**2,num1)

print(f"NEW LIST-{list(square)}")

#Example filter()

num2=[6,7,8,9,10]
print("OLD LIST-",num2)
odd_num=filter(lambda x: x%2!=0,num2)

print(f"NEW LIST-{list(odd_num)}")

#Example reduce()

from functools import reduce

num3=[1,2,3,4,5]
print("LIST-",num3)
sum=reduce(lambda x,y: x+y,num3)

print("REDUCED-",sum)
```

```python
def list_add():
  list_num=[]
  while True:
    num=input("Enter number here(if done then enter'done'):")
    if num.lower()=='done':
      break
    else:
      list_num.append(int(num))
  print(list_num)
  even_num=filter(lambda x:x%2==0,list_num)

  sum=reduce(lambda x,y:x+y,even_num)
  print(f"Sum of all even numbers in your list-{sum}")

list_add()
```

**Q2. Create a Python function that accepts a string and returns the reverse of that string.**

```python
def reverse():
  string=input("Enter a string here:")
  reverse_string=""
  for i in range(len(string)-1,-1,-1):
    reverse_string+=string[i]
  print(reverse_string)

reverse()
```

**Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def check_prime():
  num=int(input("Enter a number here:"))
  if num>=1 and num<=200:
    for i in range(2,num):
      if num%i==0:
        print(f"{num} is not a prime number")
        break
    else:
      print(f"{num} is a prime number")

check_prime()
```

**Q5. Create an iterator class in Python that generates the fibonicci numbers upto a specified number of terms.**

```python
class fibonacciiterator:
  def __init__(this,n_terms):
    this.n_terms=n_terms
    this.count=0
    this.a,this.b=0,1

  def __iter__(this):
    return this

  def __next__(this):
    if this.count >= this.n_terms:
      raise StopIteration

    if this.count==0:
      this.count += 1
      return this.a

    elif this.count==1:
      this.count += 1
      return this.b

    else:
      this.a,this.b = this.b , this.a+this.b
      this.count += 1
      return this.b

a=int(input("Enter how many fibonacci numbers you want:"))
fib=fibonacciiterator(a)
for num in fib:
  print(num,end=" ")
```

**Q7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
from google.colab import files
uploaded = files.upload()
filename=list(uploaded.keys())[0]

def read_file(filename):
  with open(filename,"r") as file:
    for i in file:
      yield i.rstrip("\n")

for i in read_file(filename):
  print(i,type(i))
```

**Q9. Write a Python program that uses 'map()' to convert a list of temperatures from Celsius to Fahrenheit.**

```python
def celsius_to_fahrenheit(celsius):
  return(celsius*9/5)+32

temperatures=[32,45,25,30,29,27]

converted_list=map(celsius_to_fahrenheit,temperatures)
fahrenheit=list(converted_list)

print(f"Temperatures in Celsius-{temperatures}\nTemperatures in Fahrenheit-{fahrenheit}")
```
